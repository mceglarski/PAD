4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
# 1

import pandas as pd
import numpy as np
import re

df = pd.read_csv('PAD04/orders.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
df
```

```python
# 2

unique_values = df['tshirt_category'].unique()
print(f' {unique_values}: {len(unique_values)}')
```

```python
# 3

def name_normalization(name, replace, replace_word):
  normalized_string = []
  for i in name.split(" "):
    if i == replace:
      normalized_string.append(replace_word)
    else:
      normalized_string.append(i)
  return " ".join(normalized_string)

df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalization(x, 'Wh', 'White'))
df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalization(x, 'Tshirt', 'T-Shirt'))
df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalization(x, 'Bl', 'Black'))
df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalization(x, 'Tennis', 'T'))
df.tshirt_category.unique()
```

```python
# 4

def tshirt_split(name, match):
  for i in name.split():
    if i in match:
      return i
  return 'unknown'

tshirt_types = ("T-Shirt", "Shirt", "Hoodie")
tshirt_colours = ("Black", "White")
tshirt_sizes = ("M", "F")

df["tshirt_type"] = df['tshirt_category'].apply(lambda x: tshirt_split(x, tshirt_types))
df["tshirt_size"] = df['tshirt_category'].apply(lambda x: tshirt_split(x, tshirt_sizes))
df["tshirt_colour"] = df['tshirt_category'].apply(lambda x: tshirt_split(x, tshirt_colours))
df
```
