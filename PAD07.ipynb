**Zadanie 1**

Przeprowadź test T studenta na danych w pliku wyniki.csv
Hipoteza zerowa brzmi: nie ma istotnej różnicy w Ocenie końcowej pomiędzy uczniami i uczennicami.
Hipoteza alternatywna brzmi: istnieje istotna różnica (bez wskazania kierunku) w Ocenie końcowej pomiędzy uczniami i uczennicami.
Przyjmij significance level (alpha) na poziomie 0.05. Czy uda się odrzucić hipotezę zerową?


```python
import pandas as pd
import plotly.express as px
import scipy.stats

df = pd.read_csv(f"PAD07/wyniki.csv", sep=',')
df
```

```python
df.groupby('plec')['ocena_koncowa'].describe()
```

```python
scipy.stats.ttest_ind(
    df[df['plec'] == 'M']['ocena_koncowa'],
    df[df['plec'] == 'F']['ocena_koncowa'],
    equal_var = True,
    alternative = 'two-sided'
)
     
```

Odpowiedź: nie można odrzucić hipotezy zerowej, ponieważ pvalue > 0.05

Korzystając z danych w pliku ZyskiFirmyX.csv stwórz model regresji liniowej i biblioteki statmodels:
•	Wyświetl wykres pokazujący zyski w czasie. 
•	Czy da się zaobserwować trend? Jaki?
•	Wyświetl podsumowanie modelu.


```python
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf
import plotly.graph_objects as go
import seaborn as sn


data2 = pd.read_csv(f'PAD07/ZyskiFirmyX.csv')
data2
```

```python
fig = px.line(data2, x='Rok', y='Zysk')
fig.update_layout(title="Wykres pokazujący zyski w czasie", xaxis_title="Rok",
    yaxis_title="Zysk")
fig.show()
```

Da się zaobserwować trend spadkowy w cenie

```python
model = smf.ols(formula="Zysk ~ Rok", data=data2).fit()
print(model.summary())
```

```python
df = pd.read_csv(f'PAD07/spendingscores.csv')
df
```

```python
replace_gender = {'Gender':{'Male':1, 'Female':0}}
df = df.replace(replace_gender)
df
```

```python
model = smf.ols(formula="SpendingScore ~ Gender + Age + AnnualIncome", data=df).fit()
print(model.summary())
```

```python
print('P Values: ', model.pvalues.values)
print("Std Errs", model.bse.values)
print('Coef: ', model.params.values)
```

```python
px.imshow(df.corr(), color_continuous_scale='Agsunset', title="Korelacja zmiennych")
```

```python
model = smf.ols(formula="SpendingScore ~ Age + AnnualIncome", data=df).fit()
print(model.summary())
```
